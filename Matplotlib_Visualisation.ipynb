```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(20,30)
```

# Scatter Plot

```python
plt.scatter(x,y,color="green",marker="+")
```

```python
y=y*y
```

# Plot

```python
plt.plot(x,y,"b+--",markersize="12")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("2D-Graph")
```

# Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,"r+--")
plt.subplot(2,2,2)
plt.plot(x,y,"b+--")
plt.subplot(2,2,3)
plt.plot(x,y,"g+--")
plt.subplot(2,2,4)
plt.plot(x,y,"y+--")
```

# Bar Graph

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

x3=[5,6]
y3=[5,9]

plt.bar(x,y)
plt.bar(x2,y2,color="g")
plt.bar(x3,y3,color="orange")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bar Graph")
```

# Histograms

```python
x=[2,3,17,9,10,15,23,25,27,33,45,53]
plt.hist(x)
plt.title("Histograms")
```

# Box PLot Using Matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

# Pie Chart

```python
sizes=[215,130,245,210,78]
explode=(0.1,0,0,0,0)
labels=["Python","Java","C++","Ruby","C"]
colors=["yellow","blue","green","violet","lightcoral"]#%0.1f%%
plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct="%0.1f%%",shadow=True)
```
