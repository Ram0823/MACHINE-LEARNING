```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Ram Baldotra\Downloads\titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.Age.median(),inplace=True)
```

```python
df.drop(["PassengerId","Name","SibSp","Ticket","Cabin","Embarked"],axis="columns",inplace=True)
```

```python
df
```

```python
target=df.Survived
```

```python
columns=pd.get_dummies(df.Sex,drop_first=True)
```

```python
df=pd.concat([df,columns],axis="columns")
df.head()
```

```python
X=df.drop(["Survived","Sex"],axis="columns")
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,target,train_size=0.75)
```

```python
from sklearn.naive_bayes import GaussianNB
naive=GaussianNB()
naive.fit(X_train,Y_train)
```

```python
pred=naive.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(Y_test,pred))
print(confusion_matrix(Y_test,pred))
accuracy_score(Y_test,pred)
```
