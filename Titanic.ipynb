```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv")
df
```

```python
new=df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns")
new.head()
```

```python
target=new.Survived
```

```python
final=new.drop(["Survived"],axis="columns")
final.head()
```

```python
list(final.Sex.unique())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex=LabelEncoder()
```

```python
final["Sex_n"]=le_Sex.fit_transform(final["Sex"])
final
```

```python
resulted=final.drop(["Sex"],axis="columns")
resulted
```

```python
resulted.isnull().sum()
```

```python
import math as m
mean=m.floor(resulted.Age.mean())
mean
```

```python
resulted["Age"]=resulted.Age.fillna(mean)
```

```python
resulted.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(resulted,target,test_size=0.3)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
