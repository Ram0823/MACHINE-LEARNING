```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/spam.csv")
df.head()
```

```python
df.Category.value_counts()
```

```python
df["Spam"]=df.Category.apply(lambda x:1  if x=="spam" else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.Message,df.Spam,train_size=0.75)
```

```python
x_train
```

## Similarly we have to convert our messages into numerical values for this wew will apply CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer()
x_train_count=CV.fit_transform(x_train.values)
```

```python
x_train_count.toarray()
```

```python
print(CV.get_feature_names())
```

```python
from sklearn.naive_bayes import MultinomialNB
mb=MultinomialNB()
mb.fit(x_train_count,y_train)
```

```python
emails=["yaar mera phone charging hai..light ke jaane ka chance hai tbhi",
        "Invitation: [Last Chance]: Newton School guarantees Tech Job of 5-40  @ Sat May 22, 2021 8pm - 8:15pm ",
    "Free entry in 2 a wkly comp to win FA Cup final"
]
```

```python
test=CV.transform(emails)
mb.predict(test)
```

```python
mb.score(CV.transform(x_test),y_test)
```
